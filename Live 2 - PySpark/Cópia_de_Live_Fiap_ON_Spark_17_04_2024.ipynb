## **Vamos falar sobre Apache Spark**

image.png

#### **O que é Apache Spark?**

O Spark é um mecanismo de análise e processamento de dados unificado que visa oferecer maior poder de processamento de dados em larga escala com módulos integrados para SQL, Streaming, Machine Learning e Processamento de gráficos.

O Spark pode ser executado no Apache Hadoop, Apache Mesos, Kubernetes, individualmente em instalações local ou em ambientes de Nuvem.

O Spark possui uma arquitetura de processamento em memória, muito flexível (e mais caro) que utilizar o MapReduce por exemplo. Isso quer dizer que MapReduce e Spark são a mesma coisa? NÂO!! Mas ambos estão entre os sistemas distribuidos mais proeminentes do mercado atual. Ambos são projetos de alto nível e que costumam ser usados juntos.

O Hadoop é usado primariamente em operações de uso intenso de disco através do MapReduce. Já o Spark fornece um processamento mais ágil por utilizar memória RAM para seu processamento.

A documentação oficial esta neste link https://spark.apache.org/docs/latest/index.html
---



#### **Visão Geral do Ecossistema Spark**

O Ecossistema Spark é composto por cinco componentes principais, veja os a seguir:


image.png



1.   *Spark Core* - Mecanismo de processamento de dados distribuido de propósitos gerais. Além disso, há bibliotecas para SQL, processamento real time (stream), machine learning e computação gráfica, e todas essas bibliotecas podem ser usadas em conjunto em uma única aplicação Spark. O Spark Core é a base de todo projeto, fornecendo despacho distribuido de tasks, programação e funcionalidades básicas de I/O.


2.   *Spark SQL* - Módulo Spark para trabalhar com dados estruturados que oferece suporte a uma maneira de acessar uma variedade de fontes de dados de maneira mais *Friendly*. Com Spark SQL voce consegue consultar dados estruturados dentro de programas Spark, usando SQL ou uma API DaraFrame.
Com Spark SQL voce pode usar sintaxe HiveQL e consultar dados existentes em armazenamentos Apache Hive.   


3. *Spark Streaming* - Módulo para criação de soluções streaming escalonáveis e com tolerancia a falhas. Através da API integrada da linguagem Spark, podemos criar jobs de streaming da mesma forma que criamos jobs em batch. Com Spark Stream podemos usar linguagens Python, Java e Scala.



4. *MLlib* - Biblioteca de machine learning escalável do Spark, com ferramentas que agregam valor ao treinamento de modelos de machine learning tornando o desenvolvimento mais fácil e escalável. MLlib contém diversos algoritmos de machine learning incluindo regressões, classificação, recomendações e clustering. Também possui pipeline de machine learning, feature engineering, avaliação de modelos, algebra linerar distribuida e diversos modelos estatísticos.  


5. *GraphX* - GraphX é a API Spark para criação de gráficos e computação paralela a gráficos. Bastante útil em analise exploratória, podemos extrair insights, transformar dados, carregar diversas fontes de dados e aplicar algoritmos de gráficos com alta flexibilidade, alto desempenho e ainda mantendo a flexibilidade, tolerância a falhas e facilidade do Spark.

---


#### **Quais são os benefícios de usar Spark?**

*Velocidade* - Podemos executar cargas de trabalho até 100 vezes mais rápidas se comparado com Apache Hadoop MapReduce. O Spark fornece alto desempenho para dados em batch e de streaming.

*Fácilidade de uso* - O Spark oferece mais de 80 operadores de alto nível que facilita a criação de apps paralelos. Podemos usar o Spark a partir dos shells do Scala, Python, R e SQL para criar aplicações rapidamente.

*Generalidade* - O Spark possibilita o uso de diversas bibliotecas incluindo SQL e DataFrames, MLlib, GraphX e Spark Streaming, possibilitando a combinação dessas bibliotecas juntas para uma mesma aplicação Spark.

*Open Source* - Por ser um projeto open source, o Spark conta com o poder coletivo uma comunidade oferecendo ideias, desenvolvimento mais ágil, coluções imediata de problemas e um tempo de lançamento mais rápido que o trabalho em equipes internas de soluções proprietárias.

---

#### **Então, porque Spark SQL e não apenas SQL?**

O Spark é um mecanismo de computação em cluster, que pode ser implantado em um cluster Hadoop ou de modo independente. O Spark faz o processamento em memória e possui as diversas vantagens que já mencionamos acima. Já um mecanismo SQL como Apache Impala, Apache Hive ou Apache Drill, não contamos com processamento em memória, biblitecas Spark para Machine Learning, diversas linguagens para apoio.

---

#### **Para quem o Spark é destinado?**

O grande desafio atual é lidar com um grande volume de dados, seja ele para as tarefas de engenheiro de dados, analista de dados ou cientista de dados. Como o Spark é uma ferramenta bastante completa e democrática, engenheiros de dados podem criar rotinas de processamento de dados usando seu alto poder computacional, bem como analistas de dados podem extrair insights rapido e facilmente usando Spark SQL e cientistas de dados treinam modelos com alta performance usando Spark MLlib.










### **Spark DataFrame X Pandas DataFrame**

O que é um dataframe? Um conjunto de dados tabulares (linhas e colunas) o conceito de dataframe nunca mudou em nenhuma linguagem de programação ou ferramenta, mas se tratando de Spark dataframe e Pandas dataframe eles são bem diferentes.


image.png


**Pandas DataFrame**

Pandas DataFrame é uma estrutura de dados tabular bidimensional potencialmente heterogênea com tamanho mutável e eixos rotulados (linhas e colunas). Os dados, linhas e colunas são os três componentes principais de um Pandas DataFrame.

**Vantagens:**

Pandas Dataframe capaz de manipular dados, como indexar, renomear, classificar e mesclar quadros de dados.

Atualizar, adicionar e excluir colunas é muito mais fácil usando o Pandas.
Pandas Dataframe suporta vários formatos de arquivo
O tempo de processamento é muito alto devido à função integrada.


**Desvantagens:**
A manipulação torna-se complexa enquanto usamos um conjunto de dados enorme.

O tempo de processamento pode ser lento durante a manipulação.


**Spark DataFrame**
No Spark, DataFrames são coleções de dados distribuídas organizadas em linhas e colunas. Cada coluna em um DataFrame recebe um nome e um tipo.

**Vantagens**

Spark carrega API fácil de usar para operar grandes conjuntos de dados.

Ele não suporta apenas "MAP" e "reduce", aprendizado de máquina (ML), algoritmos gráficos, streaming de dados, consultas SQL, etc.

Spark usa memória (RAM) para computação.

Oferece 80 operadores de alto nível para desenvolver aplicações paralelas.

**Desvantagens:**

Nenhum processo de otimização automática

Poucos algoritmos

Problema de arquivos pequenos

```python
# MLlib machine Learning com Spark
# Vector Assemble - Transforma multiplas colunas em uma coluna tipo vetor
```

```python
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.regression import LinearRegression
```

```python
assembler = VectorAssembler(inputCols=['Cylinders', 'Displacement', 'Horsepower', 'Model', 'Acceleration', 'Weight'],
                            outputCol='Features')
output = assembler.transform(df)
```

```python
final_data = output.select('Features', 'MPG')
final_data.show(4)
```

```python
train_data, test_data = final_data.randomSplit([0.8, 0.2]) # 805 dos dados para treino e 20% para teste
```

```python
model_reg = LinearRegression(featuresCol='Features', labelCol='MPG')
```

```python
model_reg = model_reg.fit(train_data)
```

```python
pred = model_reg.evaluate(test_data)
```

```python
pred.predictions.show()
```

```python
from pyspark.ml.evaluation import RegressionEvaluator
```

```python
eval = RegressionEvaluator(labelCol="MPG", predictionCol="prediction", metricName="rmse")

# Root Mean Square Error
rmse = eval.evaluate(pred.predictions)
print("RMSE: %.2f" % rmse)

# Mean Square Error
mse = eval.evaluate(pred.predictions, {eval.metricName: "mse"})
print("MSE: %.2f" % mse)

# Mean Absolute Error
mae = eval.evaluate(pred.predictions, {eval.metricName: "mae"})
print("MAE: %.2f" % mae)

# r2 - coefficient of determination
r2 = eval.evaluate(pred.predictions, {eval.metricName: "r2"})
print("r2: %.2f" %r2)
```

```python
from sklearn.datasets import load_iris
import numpy as np
import pandas as pd
from pyspark.ml.feature import StringIndexer
```

```python
data = load_iris()
data
```

```python
cols = [i.replace('(cm)','').strip().replace(' ','_') for i in data.feature_names] + ['label']

iris_df = pd.DataFrame(np.c_[data.data, data.target], columns=cols)
iris_df.iteritems = iris_df.items
```

```python
iris_df.head(5)
```

```python
df = spark.createDataFrame(iris_df)
df.show(5)
```

```python
vectorAssembler = VectorAssembler(inputCols=['sepal_length',
                                             'sepal_width',
                                             'petal_length',
                                             'petal_width'],
                                  outputCol='features')
```

```python
df_transform = vectorAssembler.transform(df)
df_final = df_transform.select(['features', 'label'])
df_final.show(5)
```

```python
from pyspark.ml.classification import NaiveBayes, MultilayerPerceptronClassifier, RandomForestClassifier
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
```

```python
# separa treino e teste
split = df_final.randomSplit([0.8, 0.2])
train = split[0]
test = split[1]
```

```python
layers = [4,5,8,12]
mlp_model = MultilayerPerceptronClassifier(layers=layers)
mlp_model = mlp_model.fit(train)
```

```python
mlp_pred = mlp_model.transform(test)
mlp_eval = MulticlassClassificationEvaluator(metricName='accuracy')
mlp_eval.evaluate(mlp_pred)
```
